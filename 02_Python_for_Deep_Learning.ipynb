## 1.4 Matplotlib Introduction

image.png

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams['figure.figsize'] = [10,3] # figure size
plt.rcParams['figure.dpi'] = 200 # figure quality (dpi = dot per inch)

data = pd.read_csv('/content/IMDB-Movie-Data.csv', nrows=50)
data.head()
print(data.columns)
plt.figure()
plt.plot(data['Rank'], data['Rating'])
plt.xlabel('Rank')
plt.ylabel('Rating')
plt.title('Rank vs Rating')
plt.show()
```

```python
#Histogram
plt.hist(data['Rating'], bins=20, edgecolor='black', color='orange')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Histogram Plot')
plt.show()
```

```python
data.columns
```

```python
plt.scatter(data['Votes'], data['Rating'], color='r')
plt.xlabel('Votes')
plt.ylabel('Rating')
plt.title('Scatter Plot')
plt.show()
```

```python
plt.boxplot(data['Rating'])
plt.show()
```

```python
data.head()
genres = data.groupby('Genre')['Rating'].mean().sort_values(ascending=False)
plt.bar(genres.index[:10], genres.values[:10], color='orange')
plt.xticks(rotation=90)
plt.show()
```

## 1.5 Introduction to Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = sns.load_dataset('iris')
print(data.columns)
# histogram
sns.histplot(x='sepal_length', data=data, kde=True, bins=20, hue='species')
plt.xlabel('Sepal Length in cm')
plt.ylabel('Count')
plt.title('Histogram of Sepal Length in Iris Dataset')
plt.show()
```

```python
sns.pairplot(data, hue='species', height=1, aspect=2)
```

```python
sns.boxplot(x='species', y = 'petal_length', data=data)
```

```python
sns.swarmplot(x='species', y = 'petal_length', data=data)
```

```python
sns.boxenplot(x='species', y = 'petal_length', data=data)
```
